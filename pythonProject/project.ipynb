```python
import pandas as pd

df = pd.read_csv("telefony.csv")
```

```python
df.head(n=10)
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt

for col in df.columns:
    fig, ax = plt.subplots(figsize=(10,8))
    ax.hist(df[col], bins=30)
    ax.set_title(col)
```

```python
for col in df.columns:
    fig, ax = plt.subplots(figsize=(10,8))
    for label in df["price_range"].unique():
        cl = df.loc[df["price_range"] == label, col]
        ax.hist(cl, bins=30, alpha=0.7, label=str(label))
    ax.legend()
    ax.set_title(col)
```

```python
binary_features = ["blue", "dual_sim", "four_g", "three_g", "touch_screen", "wifi"]
for feature in binary_features:
    df[feature] = df[feature].astype(int)

df.head(n=10)
df.info()
```

```python
continuous_features = ["battery_power", "clock_speed", "fc", "int_memory", "m_dep", "mobile_wt", "n_cores", "pc", "px_height", "px_width", "ram", "sc_h", "sc_w", "talk_time"]
for feature in continuous_features:
    fig, ax = plt.subplots(figsize=(10,8))
    for label in df["price_range"].unique():
        cl = df.loc[df["price_range"] == label, feature]
        ax.hist(cl, bins=30, alpha=0.7, label=str(label))
    ax.legend()
    ax.set_title(feature)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[continuous_features] = scaler.fit_transform(df[continuous_features])

df.head(n=10)
```

```python
for feature in continuous_features:
    fig, ax = plt.subplots(figsize=(10,8))
    for label in df["price_range"].unique():
        cl = df.loc[df["price_range"] == label, feature]
        ax.hist(cl, bins=30, alpha=0.7, label=str(label))
    ax.legend()
    ax.set_title(feature)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop("price_range", axis=1).values
y = df["price_range"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=10)
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='rbf')
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=0)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def print_metrics(y_test, y_pred):
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='macro'))
    print("Recall:", recall_score(y_test, y_pred, average='macro'))
    print("F1 Score:", f1_score(y_test, y_pred, average='macro'))

print("K-Nearest Neighbors Metrics:")
print_metrics(y_test, y_pred_knn)

print("Support Vector Machine Metrics:")
print_metrics(y_test, y_pred_svm)

print("Random Forest Metrics:")
print_metrics(y_test, y_pred_rf)
```
